```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
import numpy as np
```

```python
import face_recognition
```

```python
img1 = face_recognition.load_image_file('/content/Jiya.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1test = face_recognition.load_image_file('/content/jiya2.jpg')
img1test =cv2.cvtColor(img1test, cv2.COLOR_BGR2RGB)
```

```python
face = face_recognition.face_locations(img1)[0]
```

```python
print(face)
```

```python
encodeface = face_recognition.face_encodings(img1)[0]
```

```python
cv2.rectangle(img1,  (int(face[3]), int(face[0])), (int(face[1]),int(face[2])),(0,255,0),3)
```

```python
facetest = face_recognition.face_locations(img1test)[0]
```

```python
encodetestface = face_recognition.face_encodings(img1test)[0]
```

```python
cv2.rectangle(img1test,  (int(facetest[3]), int(facetest[0])), (int(facetest[1]),int(facetest[2])),(255,0,0),3)
```

```python
result = face_recognition.compare_faces([encodeface],encodetestface)
face_distance = face_recognition.face_distance([encodeface],encodetestface)
```

```python
print(result,face_distance)
```

```python
cv2.putText(img1test, f"{result} {round(face_distance[0],2)}",(50,50), cv2.FONT_HERSHEY_COMPLEX, 1,(0,0,255),2)
```

```python
cv2_imshow(img1)
cv2_imshow(img1test)
cv2.waitKey()
cv2.destroyAllWindows()
```
